# Merging and Visualizing  

```python
#we will merge the dataset by preforming an series of inner joins on 
#the dataframe cab_data_df, trans_id_df, customer_df and city_df based on matching columns

# Join cab_data_df with trans_id_df based on 'Transaction ID'
df = cab_data_df.join(trans_id_df.set_index('Transaction ID'), on='Transaction ID', how='inner')

# Join the resulting DataFrame with customer_df based on 'Customer ID'
df = df.join(customer_df.set_index('Customer ID'), on='Customer ID', how='inner')

# Join the resulting DataFrame with city_df based on 'City'
df = df.join(city_df.set_index('City'), on='City', how='inner')

#display the dataFrame
df.head(2)
```

## Visualizing 

```python
grouped_df = df.groupby('Company')
avg_users = grouped_df['Users'].mean()

index = avg_users.index
values = avg_users.values

colors = ['pink', 'yellow']
fig, ax = plt.subplots(figsize=(7,7))
ax.pie(values, labels=index, autopct='%1.1f%%',startangle=80, colors=colors)
ax.axis('equal')

plt.title('Users Percentage Per Company', fontsize=15)
plt.show()
```

#### conclusion: yellow cab has more users compared to the pink cab

```python
import matplotlib.pyplot as plt
import seaborn as sns

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(20, 7))

pink_df = df[df['Company'] == 'Pink Cab']
sns.lineplot(x='KM Travelled', y='Price Charged', data=pink_df, ax=ax1)
ax1.set_title('Price vs Distance in Pink Cab', fontsize=20)
ax1.set_xlabel('Distance Travelled (km)', fontsize=15)
ax1.set_ylabel('Price Charged (USD)', fontsize=15)
ax1.tick_params(axis='x', labelrotation=0, labelsize=12)

yellow_df = df[df['Company'] == 'Yellow Cab']
sns.lineplot(x='KM Travelled', y='Price Charged', data=yellow_df, ax=ax2)
ax2.set_title('Price vs Distance in Yellow Cab', fontsize=20)
ax2.set_xlabel('Distance Travelled (km)', fontsize=15)
ax2.set_ylabel('Price Charged (USD)', fontsize=15)
ax2.tick_params(axis='x', labelrotation=0, labelsize=12)

plt.show()
```

#### conclusion: there's no noticable difference in price per distance for the two cab companies

```python
pink_df = df[df['Company'] == 'Pink Cab']
pink_df = pink_df.groupby(pink_df['Date of Travel'].dt.year)
price_charged_pink = pink_df['Price Charged'].mean()
cost_trip_pink = pink_df['Cost of Trip'].mean()
profit_pink = price_charged_pink - cost_trip_pink

yellow_df = df[df['Company'] == 'Yellow Cab']
yellow_df = yellow_df.groupby(yellow_df['Date of Travel'].dt.year)
price_charged_yellow = yellow_df['Price Charged'].mean()
cost_trip_yellow = yellow_df['Cost of Trip'].mean()
profit_yellow = price_charged_yellow - cost_trip_yellow

fig, ax = plt.subplots(figsize=(10, 7))

profit_pink.plot(kind='bar', ax=ax, position=0, width=0.4, align='edge', color = 'pink')
profit_yellow.plot(kind='bar', ax=ax, position=1, width=0.4, align='edge', color = 'yellow')

ax.set_title('Profit per Year for Pink Cab and Yellow Cab', fontsize=20)
ax.set_xlabel('Year', fontsize=15)
ax.set_ylabel('Profit (USD)', fontsize=15)
ax.tick_params(axis='both', labelsize=12)
ax.legend(['Pink Cab', 'Yellow Cab'], fontsize=12)

plt.show()
```

#### conclusion: the yellow cab makes significantly more profit per year in comparsion to pink cab

```python
# Calculate profit per city for Pink Cab
pink_profit = df[df['Company'] == 'Pink Cab'].groupby(['City'])[['Price Charged', 'Cost of Trip']].mean()
pink_profit['Profit'] = pink_profit['Price Charged'] - pink_profit['Cost of Trip']

# Calculate profit per city for Yellow Cab
yellow_profit = df[df['Company'] == 'Yellow Cab'].groupby(['City'])[['Price Charged', 'Cost of Trip']].mean()
yellow_profit['Profit'] = yellow_profit['Price Charged'] - yellow_profit['Cost of Trip']

# Combine the profit data for both companies
profit = pd.concat([pink_profit, yellow_profit], axis=1)
profit.columns = ['Pink Cab Price Charged', 'Pink Cab Cost of Trip', 'Pink Cab Profit', 'Yellow Cab Price Charged', 'Yellow Cab Cost of Trip', 'Yellow Cab Profit']
profit = profit.loc[:, ['Pink Cab Profit', 'Yellow Cab Profit']]
profit = profit.sort_values(by='Pink Cab Profit', ascending=False)

# Plot the bar chart
fig, ax = plt.subplots(figsize=(10, 7))
profit.plot(kind='bar', ax=ax, color=['pink', 'yellow'], width=0.4)

ax.set_title('Profit per City for Pink Cab and Yellow Cab', fontsize=20)
ax.set_xlabel('City', fontsize=15)
ax.set_ylabel('Profit (USD)', fontsize=15)
ax.tick_params(axis='both', labelsize=12)
ax.legend(['Pink Cab', 'Yellow Cab'], fontsize=12)

plt.show()
```

#### conclusion: the yellow cab company makes more profit across the cites in the US than the pink cab company

```python
# Group the data by gender and count the number of rides
gender_counts = df.groupby('Gender')['Transaction ID'].count()

# Create a bar chart of the gender counts
gender_counts.plot(kind='bar', color=['pink', 'blue'], alpha=0.5)

# Add labels and title to the plot
plt.xlabel('Gender')
plt.ylabel('Number of Rides')
plt.title('Number of Rides by Gender')

# Show the plot
plt.show()
```

#### conclusion: males are makeup more of the customers for cab companies in comparsion to females

```python
# Filter the data to only include customer ID, income, transaction ID, and Pink Cab rides
df = df[['Customer ID', 'Income (USD/Month)', 'Transaction ID']]

# Group the data by customer ID and count the number of transactions for each customer
grouped_data = df.groupby('Customer ID').count().reset_index()

# Create a line graph of number of transactions vs. income
plt.plot(grouped_data['Transaction ID'], grouped_data['Income (USD/Month)'])

# Add labels and title to the plot
plt.xlabel('Number of Transactions')
plt.ylabel('Income (USD/Month)')
plt.title('Relationship between Frequency of Cab Use and Income')

# Show the plot
plt.show()
```

#### conclusion: people with higher incomes tend to use cab more often than those who don't
